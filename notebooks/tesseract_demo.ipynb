```python
import PIL
import pytesseract
import os
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image, display
from skimage.filters import threshold_otsu
from imageio import imsave


path = '.'

converted_image_dir = "converted"
if not os.path.exists(converted_image_dir):
    os.makedirs(converted_image_dir)
    
for file in os.listdir(path):
    if (".png" not in file):
        continue
    image = PIL.Image.open(file)
    print(pytesseract.image_to_string(image, config=conf))
    display(Image(filename=file))
    greyscale = image.convert('L')
#        display(greyscale)
    thres = threshold_otsu(np.asarray(greyscale))
    img = greyscale > thres
    converted_file = os.path.join(converted_image_dir, file[:-4] + "_bw.png")
    imsave(converted_file, img.astype(int))
    print(pytesseract.image_to_string(PIL.Image.open(converted_file)))
    display(Image(filename=converted_file))
        
```
